```python
import requests
from bs4 import BeautifulSoup

URL = "https://eqvista.com/fortune-500-companies-in-the-us/"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")
```

```python
import pandas as pd

df = pd.DataFrame(columns=[column_name.text.strip() for column_name in soup.find_all("th")])

table_row = soup.find_all("tr")[1:]
table_row_data = [[data.text.strip() for data in row.find_all("td")] for row in table_row]

for row in table_row_data:
    df.loc[len(df)] = row

display(df)
df.to_csv("csv_files/top_100_companies.csv", index=False)
    


```
